```python
# demo1
import numpy 
import matplotlib.pyplot as plt

# 定义目标函数 g(x)
def g(x):
    return numpy.tan(x - 1)

# RM算法参数
x = 3  # 初始值
epsilon = 1e-6  # 终止条件
max_iterations = 100  # 最大迭代次数
history = []  # 用于存储每次迭代后的 x 值
y = []
for n in range(1, max_iterations + 1):
    step_size = 1 / n  # 步长
    x_next = x - step_size * g(x)  # 更新 x 值
    history.append(x_next)
    y.append(g(x))

    if abs(g(x_next)) < epsilon:
        break

    x = x_next  # 更新 x

# 可视化收敛过程
plt.plot(history, marker='o')
plt.xlabel('Iteration')
plt.ylabel('x value')
plt.title('Convergence of RM Algorithm')
plt.grid(True)
plt.show()
```
